```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
def encode_categorical(df,headers,encodingType):
    for header in headers:
        if encodingType == 'label': label_encoder = LabelEncoder()
        df[header] = label_encoder.fit_transform(df[header])
    return df

import numpy as np

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Cost function
def compute_cost(X, y, weights):
    m = len(y)
    h = sigmoid(np.dot(X, weights))
    cost = (-1/m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
    return cost

# Gradient function
def compute_gradient(X, y, weights):
    m = len(y)
    h = sigmoid(np.dot(X, weights))
    gradient = (1/m) * np.dot(X.T, (h - y))
    return gradient

# Training function using gradient descent
def train_logistic_regression(X, y, learning_rate, iterations):
    weights = np.zeros(X.shape[1]) #beta
    cost_history = []

    for i in range(iterations):
        gradient = compute_gradient(X, y, weights)
        weights -= learning_rate * gradient
        cost = compute_cost(X, y, weights)
        cost_history.append(cost)

        if i % 100 == 0:print(f"Iteration {i}: Cost {cost}")

    return weights, cost_history

# Prediction function
def predict(X, weights, threshold=0.5):
    probabilities = sigmoid(np.dot(X, weights))
    return probabilities >= threshold

def plot_predict(y1,y2):
    plt.figure(figsize=(20, 5))  # Optional: set the figure size

    plt.scatter(range(len(y1)), y1, color='b', label='actual')
    plt.scatter(range(len(y2)), y2, color='r', label='predicted')

    plt.title('Simple Line Plot')
    plt.xlabel('X Axis')
    plt.ylabel('Y Axis')
    plt.legend()

    plt.grid(True)  # Optional: add gridlines
    plt.tight_layout()  # Optional: improve spacing

    plt.show()
```

```python
#input data preparation
df_bank = pd.read_csv('banking.csv')

relevant_features = ['previous', 'job', 'default', 'month', 'day_of_week','poutcome','y']
df_reduced = df_bank[relevant_features]
#for feature in relevant_features: print(feature,": " ,df_reduced[feature].unique())

df_clean = encode_categorical(df_reduced.copy(),['job','default','month','day_of_week','poutcome'],encodingType='label')
df_clean.to_csv('ML_test_bank.csv',index=False)
display(df_clean)
```

```python
X = df_clean.drop('y', axis=1).to_numpy()
y = df_clean['y'].to_numpy()

#X = np.array([[2, 3,7], [1, 5,8], [4, 2,9], [3, 6,0],[1,2,5]])
#y = np.array([1,0,1,0,0])

# Adding intercept term
X = np.hstack((np.ones((X.shape[0], 1)), X))

# Training the model
learning_rate = 0.1
iterations = 2000
weights, cost_history = train_logistic_regression(X, y, learning_rate, iterations)

predictions = predict(X, weights)

print("Predictions: ", predictions)

accuracy = 0
for i in range(len(predictions)):
    flag = 0
    if predictions[i] == True: flag = 1
    if y[i] == flag: accuracy+=1

print(accuracy,len(predictions),accuracy/len(predictions))

plot_predict(y,predictions)
```
